## 1. 계산기 프로그램 만들며 클래스 알아보기

```python
result = 0 #초기값

def add(num):
    global result
    result += num
    return result

print(add(3))
print(add(4))
```

```python
# 2대의 계산기
result1 = 0
result2 = 0

def add1(num):
    global result
    result +=num
    return result

def add2(num):
    global result2
    result2 += num
    return result2
print(add1(3))
print(add1(4))
print(add2(7))
```

```python
#여러대 계산기를 클래스로 해결하기
class Calculator:
    def __init__(self):
        self.result = 0 # 초기값 실행
    def add(self,num):
        self.result *= num
        return self.result
```

```python
cal1 = Calculator()
cal2 = Calculator()
```

```python
print(cal1.add(3))
```

```python
print(cal1.add(24))
```

## 4)생성자 함수(constructor)

- 생성자는 객체가 생성될 때 자동으로 싫생되는 메서드
- 객체의 초기값 설정

```python
class FourCal:
    #생성자 함수 : 객체생성시 자동 호출
    def  __init__(self, frist, second):
        self.frist = frist
        self.second = second

    def add(self):
        result = self.frist + self.second
        return result

    def sub(self):
        result = self.frist - self.second
        return result       
        
    
    def mul(self):
        result = self.frist * self.second
        return result

    def div(self):
        result = self.frist / self.second
        return result
```

```python
a = FourCal()
```

```python
a = FourCal(4,2)
```

```python
a.add()
```

```python
a.sub()
```

```python
a.div()
```

```python
a.mul()
```

```python
a.second
```

```python
a.frist
```
