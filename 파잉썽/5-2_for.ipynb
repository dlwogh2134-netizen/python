## 3.for문의 응용

```python
marks = [90,25,67,45,80] #학생들의 시험 점수 리스트
number = 0 #학생들에게 붙여 줄 번호
for mark in marks: #90,25,67,45,80을 순서대로 amrk에 대입
    number += 1
    if mark >=60:
        print("%d번 학생은 합격입니다." %number)
    else:
        print("%d번 학생은 불합격입니다." %number)
```

## 8.리스트 컴프리헨션 사용하기

```python
a = [1,2,3,4]
result = []

for num in a:
    result.append(num * 3)
    print(result)
```

```python
# 리스트 컴프리헨션 (listcomprehension)
a [1,2,3,4]
result = [num*3 for num in a]
print(result)
```
