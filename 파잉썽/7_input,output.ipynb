## 파일 읽고 쓰기
### 1)파일 생성하기

```python
f = open('newfile.txt', 'w') #상대 경로
f.close() #파일 객체를 닫아주는 역할을 함.
```

```python
# 특정 경로 파일 생성
f = open('d:/newfile.txt', 'w') #절대 경로
f.close()
```

## 2) 파일을 쓰기 모드로 열어 내용 쓰기

```python
f = open('newfile.txt', 'w')

for i  in range(1,11):
    data = '%d번째 줄입니다.\n'%i
    f.write(data) #파일에 write

f.close()
```

```python
f = open('newfile.txt', 'w')

for i  in range(1,11):
    data = '%d번째 줄입니다.\n'%i
    print(data)
```

### 3) 파일을 읽는 여러가지 방법

- redline() 함수 이용하기

```python
f = open ('newfile.txt', 'r')
line = f.readline() #한 줄씩 읽어오는 함수
print(line)
f.close()
```

```python
f = open ('newfile.txt', 'r')
while True:
    line = f.readline()
    if not line:
        break
    print(line)
```

- 사용자의 입력을 받아 여러줄 출력

```python
while True:
    data = input()
    if not data:
        break
    print(data)
    
```

- readlines() 함수 사용하기

```python
f = open ('newfile.txt', 'r')
lines = f.readlines()
lines
```

- 줄바꿈 (\n)문자 제거:srtip()

```python
f = open ('newfile.txt', 'r')
lines = f.readlines()

for line in lines :
    line = line.strip()
    print(line)

f.close()
```

```python
f = open ("newfile.txt", "r")
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

- 파일 객체를 for문으로 읽어오기

```python
f = open('newfile.txt','r')
for line in f:
    print(line)

f.close
```

### 4)파일에 새로운 내용 추가하기

```python
f = open('newfile.txt', 'a')

for i in range(11, 21):
    data ='%d번째 줄 입니다\n'  %i
    f.write(data)

f.close()
```

### 5)with문과 함꼐 사용하기 : 파일 열고 닫는걸 자동으로 처리

```python
with open ('foo.txt','w') as f:
    f.write('life is too short, you need python')
```
