### 3. 교집합,, 합집합, 차집합 구하기


```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
```

```python
# 교집합(&) 구하기
s1 &s2
```

```python
s1 .intersection(s2)
```

```python
# 합집합 (ㅣ) 구하기
```

```python
s1 .union(s2)
```

```python
# 차집합 (-) 구하기
s1 - s2
```

```python
s2 - s1
```

```python
s1.difference(s2)
```

```python
s2.difference(s1)
```
