## while문 만들기

```python
prompt = """
1. And
2. Del
3. List
4. Quit

Enter number: """
```

```python
number = 0

while number !=4: #number 4가 아닌 동안에 반복문 실행
    print(prompt)
    number = int(input())
```

## continue문 : continue문 바로 뒤에 오는 실행문만 실행안하고 반복문으로 돌아감

```python
a = 0

while a < 10:
    a += 1
    if a % 2 == 0:
        continue
    print(a)    
```
